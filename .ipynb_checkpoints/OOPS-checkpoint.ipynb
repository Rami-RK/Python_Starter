# Example 1
* We are going to make a class named 'Arithmatic_Operation' which is capable of calculating addition, and multiplication of given three numbers.
* First Making function for each operation and calling them.

```python
# Making 'add' function for addition  
def add(a,b,c):
    result=a+b+c
    return result
#####Callig the function#####
add(3,4,5)
```

```python
# Making 'multiply' function for multiplication  
def multiply(a,b,c):
    result=a*b*c
    return result
#####Callig the function#####
multiply(3,4,5)
```

### Now implementing above operations making a class:

### Running above code and seeing the result:

```python
class Arithmetic_Operation():
    def __init__(self,a,b,c):     # Constructor, it intialiazes attributes 
        self.a=a
        self.b=b
        self.c=c
    def add(self):                # First Method of class Arithmatic_Operation
        result=self.a+self.b+self.c
        return result
    def multiply(self):           # Second Method of class Arithmatic_Operation
        result=self.a*self.b*self.c
        return result
```

```python
obj1=Arithmetic_Operation(3,4,5) 
#'obj1' is an object of class'Arithmatic_Opeation'
# Making an object using class,is defined as 'Instantiation'                  
```

##### After creating object now we can apply attributes and methods to see the result as given below:
* Applying/Accessing attributes gives the value of attributes that were passed while making object using class.
* Applying/Calling method gives the result of operation that were implemented inside particular method on attributes.

```python
print(obj1.a)                    # Applying attribute 'a' on object 
print(obj1.b)                    # Applying attribute 'b' on object
print(obj1.c)                    # Applying attribute 'c' on object
print(obj1.add())                # Applying method 'add()' on object
print(obj1.multiply())           # Applying method 'multiply()' on object
```

#### Note : Parenthesis is used at the last of method.But it is never used while applying attributes.

```python
class Human:
    
    def __init__(self, name, no_meal, sleep_hour, work_hour):
        self.name = name
        self.no_meal =no_meal
        self.sleep_hour = sleep_hour
        self.work_hour =  work_hour

    def eating(self):
        print("{}  eats {} times a day.".format(self.name, self.no_meal))
    def sleeping(self):
        print("{} sleeps {} hours a day.".format(self.name, self.sleep_hour))
    def working(self):
        print("{} works {} hours a day.".format(self.name, self.work_hour))
```

```python
# creating an object of class Human
ram = Human("Ram", 2, 7,9)
# Appplying attributes
print(ram.name)
print(ram.no_meal)
print(ram.sleep_hour)
print(ram.work_hour)
# Applying methods on object
ram.eating()
ram.sleeping()
ram.working()
```

## Example 4
* We are going to make a class named 'stat_summary' for calculating mean and standard deviation of a list of numbers.
* I this example attribute is a list i.e user have to pass a list of number.
* This class has two method one for calculating mean and another for calculating standard deviation.

```python
class stat_summary():
    def __init__(self, list1):  
        self.list1 = list1    # Initialize attributes to describe a stat_summary.
                   
    def mean(self):          # First Method for calculating mean
        count=0
        sum1=0
        for i in self.list1:
            count+=1
            sum1+=i
        mean1=sum1/count
        return mean1   
        
    def std_dev(self):       # SEcond Method for calcualting standard deviation 
        m=sum(self.list1)/len(self.list1)
        s=0
        for i in self.list1:
            s=s+(i-m)**2
        std=(s/(len(self.list1)-1))**0.5
        return std
```

```python
List=[2,1,3,6,4,5] # This is the list and we are going to calculate mean and standard deviation of it using our class.
# Making an object named-'stat_cal' using the class:
stat_cal=stat_summary(List) 
# Accessing the attribute, i.e. it will display the list we have passed in class while making object
print('List passed : ',stat_cal.list1)
# Applying methods on object
print('Mean of the list of numbers : ',stat_cal.mean())
print('Standard deviation of the list of numbers : ',stat_cal.std_dev())
```

```python
## Cross verification using statistics module
import statistics as st
print(st.mean(List))
st.stdev(List)
```

## Example 5
* In this example, we will see method can also take arguments like in general funcions. This class has two method.
* Method ->'simple_sum' -> just return the sum of three attributes passed while making object.
* Method -> 'sum_after_exp'-> calculates sum after raising each number to the power of exponent, value of which is provided by user while applying this method . 

```python
class exponent_sum():
    
    def __init__(self,x,y,z):  # constructor: User have to pass x,y,z while making object using class.
        self.x=x
        self.y=y
        self.z=z
        
    def simple_sum(self):               # First method
        return self.x+self.y+self.z
    
    def sum_after_exp(self,exponent):    # Second method
        return self.x**exponent+ self.y**exponent+self.z**exponent
    
    
 # Note : In the second method user have to pass a value for exponent while applying this method on object.
```

```python
summation=exponent_sum(2,3,4) # making object 
print( 'Simple sum of three numbers : ',summation.simple_sum()) # applying method
print('Sum after raising to the power of value of expoenent : ',summation.sum_after_exp(2)) # applying method
# Here user has passed value '2' for exponent. Note self is never passed by user.
```

##### Add a little logic to our program  > update odometer section
* In 'update_odometer', method, we have added a condition if the supplied 'mileage' is greater than existing 'odometer_reading'only then odometer_reading will be updated otherwise- "You can't roll back an odometer" msg. will be displayed.

```python
class Car():
    def __init__(self, make, model, year):
        self.make = make                          # Initialize attributes to describe a car.
        self.model = model
        self.year = year
        self.odometer_reading = 0
    def get_descriptive_name(self):               # Return a neatly formatted descriptive name.      
        long_name = str(self.year) + ' ' + self.make + ' ' + self.model
        return long_name.title()
    def read_odometer(self):
        print("This car has " + str(self.odometer_reading) + " miles on it.")

    def update_odometer(self, mileage): # Set the odometer reading to the given value.
        if mileage >= self.odometer_reading:
            self.odometer_reading = mileage
        else:
            print("You can't roll back an odometer!")
```

## Inheritance

You dont always have to start from scratch when writing a class. If the class
youre writing is a specialized version of another class you wrote, you can
use inheritance. When one class inherits from another, it automatically takes
on all the attributes and methods of the first class. The original class is
called the parent class, and the new class is the child class. The child class
inherits every attribute and method from its parent class but is also free to
define new attributes and methods of its own.

```python
############# Parent class :> ######################

class Car():         
    def __init__(self, make, model, year):
        self.make = make                          # Initialize attributes to describe a car.
        self.model = model
        self.year = year
        self.odometer_reading = 0
        
    def get_descriptive_name(self):               # Return a neatly formatted descriptive name.      
        long_name = str(self.year) + ' ' + self.make + ' ' + self.model
        return long_name.title()
    
    def read_odometer(self):
        print("This car has " + str(self.odometer_reading) + " miles on it.")

    def update_odometer(self, mileage):  # Set the odometer reading to the given value.
        self.odometer_reading = mileage
        
############## Child class :> ######################   

class ElectricCar(Car):
    def __init__(self,make,model, year):
        super().__init__(make, model, year)  #Initialize attributes of the parent class.
```

### Few points to be noted : 
* In child class we have to pass the name of Parent class in parenthesis after name of the child class.Here 'Car' is passed. 
* In child class we have not initialized attributes, like self.make=make, self.model=model and so on.
* All the attributes in child class are of parent class so these are  not inialized here.
* All the attributes which  belongs to parent class are passed to parent class  through super(). __ init __(), see above, so that these can be initialized in parent class.
* In this example,child class has no attributes of its own, so not initialized.But, it definitely can have multiple attributes of its own and that have to be initialized using self.

```python
my_ecl_car=ElectricCar('Tesla','T84',2019) # Now making object using child class.
```

```python
my_ecl_car.get_descriptive_name() # In  oject of child class, all methods of parent class are valid.
```

### Defining Attributes and Methods for the Child Class
* Once you have a child class that inherits from a parent class, you can add any new attributes and methods necessary to differentiate the child class from the parent class.

* 'describe_battery()' and 'fill_gas()' are two methods defined inside child class.
* There is no attribute for child class except for a default attribute 'self.battery_size'. 

```python
class Car():
    def __init__(self, make, model, year):
        self.make = make                          # Initialize attributes to describe a car.
        self.model = model
        self.year = year
        self.odometer_reading = 0
        
    def get_descriptive_name(self):               # Return a neatly formatted descriptive name.      
        long_name = str(self.year) + ' ' + self.make + ' ' + self.model
        return long_name.title()
    
    def read_odometer(self):
        print("This car has " + str(self.odometer_reading) + " miles on it.")

    def update_odometer(self, mileage):  # Set the odometer reading to the given value.
        self.odometer_reading = mileage
        
    def fill_gas(self):
        print('You need to fill gas tank.')
######################################################################  Child class :>    
class ElectricCar(Car):
    def __init__(self, make, model, year):
        super().__init__(make, model, year)  # Initialize attributes of the parent class.
        self.battery_size = 70    # Defining default attribute of child class
        
    def describe_battery(self,weight): # Defining Method for child class # Prinst a statement describing the battery size.
        print("This car has a " + str(self.battery_size) + "-kWh battery.")
        print('Battery weights '+str(weight) + ' kg.')
        
    def fill_gas(self): # Note:concept of over-riding parent method(explained below).
        print("Electric car doesn't have a gas tank.")
```

```python
my_ecl_car=ElectricCar('Tesla','T84',2019)
```

```python
my_ecl_car.describe_battery(69)
```

```python
my_ecl_car=ElectricCar('Tesla','T84',2019)
```

```python
my_ecl_car.fill_gas()
```

### Overriding Methods from the Parent Class
* You can override any method from the parent class that doesnt fit what youre trying to model with the child class. To do this, you define a method in the child class with the same name as the method you want to override in the parent class. Python will disregard the parent class method and only pay attention to the method you define in the child class. Say the class Car had a method called fill_gas_tank(). This method is meaningless for an all-electric vehicle, so you might want to override this method. Heres one way to do that: Already done in above example. See what happens after removing 'fill_gas' method from child class and running it.

### Instances as an Attributes
* In this example we have made three classes 'Car', 'Battery', and 'ElectricCar'.Class 'Car' is used as a parent class for   child Class i.e. for 'ElectricCar'.
* Inside class 'ElectricCar' -->  Notice this portion --> self.battery = Battery(self.bat_weight)
* R.H.S is 'Battery' class and 'self.bat_weight' is an attribute of 'ElectricCar' class which is passed as an attribute'weight' of 'Battery' class.
* L.H.S. is an object/Instance of 'Battery' class. But it is used in so clever way that it is also an attribute of 'ElectricCar' class.

```python
class Car():
    def __init__(self, make, model, year):
        self.make = make                          # Initialize attributes to describe a car.
        self.model = model
        self.year = year
        self.odometer_reading = 0
    def get_descriptive_name(self):               # Return a neatly formatted descriptive name.      
        long_name = str(self.year) + ' ' + self.make + ' ' + self.model
        return long_name.title()
    def read_odometer(self):
        print("This car has " + str(self.odometer_reading) + " miles on it.")

    def update_odometer(self, mileage):           # Set the odometer reading to the given value.
        self.odometer_reading = mileage
        
    def fill_gas(self):
        print('You need to fill gas tank.')

#############################################################################        
class Battery():
    def __init__(self,weight,battery_size=70):
        self.weight=weight
        self.battery_size = battery_size
    def describe_battery(self):
        print("This car has a " + str(self.battery_size) + "-kWh battery.")
        print('Battery weights '+str(self.weight) + ' kg.')
######################################################################  Child class :>    
class ElectricCar(Car):
    def __init__(self, make, model, year,bat_weight):
        self.bat_weight=bat_weight
        super().__init__(make, model, year)     # Initialize attributes of the parent class.
        self.battery = Battery(self.bat_weight) # Add an attribute called self.battery, and it is an object of 'Battery' class.
        
    def fill_gas(self): ## Note:concept of over-riding parent method
        print("Electric car doesn't have a gas tank.")        
```

```python
my_tesla = ElectricCar('tesla', 'model s', 2016,86) # making object
print(my_tesla.get_descriptive_name())  # applying parent's method on object
```

```python
my_tesla.battery.describe_battery()
# Here 'battery' attribute is applied on object 'my_tesla'.
# But the part 'my_tesla.battery' is also is an object or isntance of 'Battery' class.
# Thus 'describe_battery' being a method of 'Battery' class is applied on 'my_tesla.battery'. 
```
